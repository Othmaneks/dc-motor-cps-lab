# Baseline S0 — reponse au creneau, PWM 150

Donnees : `data/S0_step_001.csv` (2001 points, 100 Hz, 5 cycles ON/OFF de 2s).

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/S0_step_001.csv")
df.head()
```

```python
# 960 ticks/tour en sortie (32 fronts x reduction 30:1)
CPR = 960

df["t"] = df["t_ms"] / 1000
df["rev"] = -df["count"] / CPR   # signe inverse, cf convention cablage
```

```python
df["rpm"] = df["rev"].diff() / df["t"].diff() * 60
df["rpm_f"] = df["rpm"].rolling(9, center=True).mean()   # lissage bruit quantif
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 6), sharex=True)
ax[0].plot(df.t, df.rev)
ax[0].set_ylabel("position (tr)")
ax[0].grid(alpha=0.3)
ax[1].plot(df.t, df.rpm_f, "r")
ax[1].set_ylabel("vitesse (tr/min)")
ax[1].set_xlabel("t (s)")
ax[1].grid(alpha=0.3)
plt.tight_layout()
```

Plateaux ~335 tr/min, tres repetables (ecart <0.5% entre cycles).
Montee pilotee rapide (~0.3 s), descente en roue libre lente (~1.5 s) : deux regimes physiques differents.
